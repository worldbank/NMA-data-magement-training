## A Simple geoprocessing task
Given two datasets as follows: a polygon dataset with section boundaries and corresponding population and a point datasete of building footprints, 
generate building counts for each section. As previously mentioned, the advantage of doing this in Python as opposed to Desktop GIS is the ability to automate  and run this process again without extra effort.

````python
buildings_file = "../data/google-building-footprints-sample.shp"
section_pop_file = "../data/Section.shp"
````

````python
# Read the shapefiles using Geopandas
#gdf_bld = gpd.read_file(buildings_file)
gdf_sect = gpd.read_file(section_pop_file)
````

````python
# We gonna use SCODE as a unique ID for each section and verify that its indeed unique for each row below
assert gdf_sect.SCODE.nunique() == gdf_sect.shape[0]
````

````python
# perfom spatial join for the polygon and  point shapefiles
gdf_sect_bld = gpd.sjoin(left_df=gdf_sect, right_df=gdf_bld, how='inner', predicate='contains')
````

````python
# count points in polygons by using groupby
gdf_sect_bld_grp = gdf_sect_bld.groupby(['SCODE']).size().reset_index()
````

````python
# add better namee for building count and re-merge with the Section shapefile
gdf_sect_bld_grp.rename(columns={0: 'bld_cnt'}, inplace=True)
gdf_sect_bld = gdf_sect.merge(gdf_sect_bld_grp, on='SCODE', how='inner')
````

````python
# Ensure everything matched perfectly 
assert gdf_sect.shape[0] == gdf_sect_bld.shape[0] 
````

### And lets do some sanity check

````python
corr = pd.DataFrame.corr(gdf_sect_bld[['SUM_TOTPOP','bld_cnt']]) 
````

````python
corr_text = corr.iloc[0]['bld_cnt']
corr_text
````

## Interactive Visualisations with Bokeh

The simple interactive example below has been copied from [Bokeh Documentation] (https://demo.bokehplots.com/apps/moviesfrom) for demonstration.

````python
from bokeh.plotting import figure, show, output_file
import numpy as np
N = 4000
x = np.random.random(size=N) * 100
y = np.random.random(size=N) * 100
radii = np.random.random(size=N) * 1.5
colors = [
    "#%02x%02x%02x" % (int(r), int(g), 150) for r, g in zip(50+2*x, 30+2*y)
]

TOOLS="hover,crosshair,pan,wheel_zoom,zoom_in,zoom_out,box_zoom,undo,redo,reset,tap,save,box_select,poly_select,lasso_select,"

p = figure(tools=TOOLS)

p.scatter(x, y, radius=radii,
          fill_color=colors, fill_alpha=0.6,
          line_color=None)

output_file("color_scatter.html", title="color_scatter.py example")

# please inspect the plot in the browser
show(p)
````

### Explore most prevalent words in a document using wordcloud

````python
def convert_pdf_pages_into_text(the_document):
    """
    Loads a PDF document and converts each page content into text string
    """
    # Get text from all pages and put in a list
    reader = PdfReader(the_document)
    number_of_pages = len(reader.pages)
    page_texts = []
    for i in range(number_of_pages):
        try:
            page = reader.pages[i]
            text = page.extract_text()
            text = re.sub(r'==.*?==+', '', text)
            text = text.replace('\n', '')
            page_texts.append(text)
        except:
            pass
    
    # Combine all pages text into one text string
    text = ",".join(page_texts)
    
    return text
````

````python
# Define a function to plot word cloud
def plot_cloud(wordcloud):
    # Set figure size
    plt.figure(figsize=(40, 30))
    # Display image
    plt.imshow(wordcloud) 
    # No axis details
    plt.axis("off")
````

#### Create a wordcloud

````python
# Load text from the document
transparency_document = "../data/documents-as-data/Mainstreaming Transparency.pdf"
text = convert_pdf_pages_into_text(transparency_document)


# Create a WordCloud object
wordcloud = WordCloud(width= 2000, height = 1000, random_state=1, background_color='salmon', colormap='Pastel1', 
                      collocations=False, stopwords = STOPWORDS, max_words=100)

# Generate a word cloud
wordcloud.generate(text)

# Plot the wordcloud
plot_cloud(wordcloud)
````

### Explore topics in a document a.k.a topic modelling
Given a collection of documents or a single document, explore the prevalent topics within that document or the collection of documents.

````python
# Install PyPDF2 if its not available
try:
    import pickle 
    import gensim
    import pyLDAvis.gensim_models
    import pyLDAvis
    import gensim.corpora as corpora
    from gensim.utils import simple_preprocess
    import nltk
    from nltk.tokenize import TweetTokenizer, sent_tokenize, RegexpTokenizer
    nltk.download('stopwords')
    from nltk.corpus import stopwords
    from pprint import pprint
except:
    !pip install gensim
    !pip install nltk
    !pip install pyLDAvis
````

### preprocess the data/text

````python
def remove_stopwords(texts):
    words = []
    for word in texts:
        if word in stopwords:
            continue
        else:
            words.append(word)
    return words
````

````python
transparency_document = "../data/documents-as-data/Mainstreaming Transparency.pdf"
text = convert_pdf_pages_into_text(transparency_document)
text
````

````python
tokenizer = RegexpTokenizer(r'\w+')
stop = list(set(stopwords.words('english')))
data_words = [i.lower() for i in tokenizer.tokenize(text) if i not in stop]
data_words
````

````python
# Create Dictionary
id2word = corpora.Dictionary([data_words])
# Create Corpus
texts = data_words
# Term Document Frequency
corpus = [id2word.doc2bow([text]) for text in texts]
# View
print(corpus[:1][0][:30])
````

````python
# number of topics
num_topics = 10
# Build LDA model
lda_model = gensim.models.LdaMulticore(corpus=corpus,
                                       id2word=id2word,
                                       num_topics=num_topics)
# Print the Keyword in the 10 topics
pprint(lda_model.print_topics())
doc_lda = lda_model[corpus]
````

````python
# Visualize the topics
pyLDAvis.enable_notebook()
LDAvis_data_filepath = 'ldavis_prepared_{}'.format(num_topics)
# # this is a bit time consuming - make the if statement True
# # if you want to execute visualization prep yourself
if 1 == 1:
    LDAvis_prepared = pyLDAvis.gensim_models.prepare(lda_model, corpus, id2word)
    with open(LDAvis_data_filepath, 'wb') as f:
        pickle.dump(LDAvis_prepared, f)
# load the pre-prepared pyLDAvis data from disk
with open(LDAvis_data_filepath, 'rb') as f:
    LDAvis_prepared = pickle.load(f)
pyLDAvis.save_html(LDAvis_prepared, 'ldavis_prepared_{}.html'.format(num_topics))
LDAvis_prepared
````
